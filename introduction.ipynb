NODES AND EDGES

```python
from langgraph.graph import StateGraph      # used to build the workflow (graph)
from langgraph.graph import END             # tells the graph to stop execution here
from typing import TypedDict                # used to define the structure of the state
from langchain.chat_models import init_chat_model
```

```python
# 1. Define state (acts as global object passed between nodes)
class State(TypedDict):
    input: str          # user question
    output: str         # llm answer
```

```python
# 2. LLM
llm = init_chat_model("google_genai:gemini-2.5-flash-lite")
```

```python
# 3. Node function
def chatbot(state: State):                      # recieves current state as input
    response = llm.invoke(state["input"])       # takes user input from state and passes it to llm
    return {"output": response.content}         # returns the updated state of output that includes the llm response
```

```python
# 4. Build graph
graph = StateGraph(State)                       # create a graph object and pass the state structure
graph.add_node("chatbot", chatbot)              # add "chatbot" node and pass the user defined chatbot function 
graph.set_entry_point("chatbot")                # define where execution starts from
graph.add_edge("chatbot", END)                  # add edge from chatbot node to end
```

```python
# 5. Compile
app = graph.compile()                            # convert graph into an executable app
```

```python
# 6. Run
result = app.invoke({"input": "What is AI?"})
print(result["output"])
```
